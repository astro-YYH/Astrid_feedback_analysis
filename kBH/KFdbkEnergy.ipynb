```python
# plot feedback energy vs. BH mass

fig, ax = plt.subplots(figsize=(5, 4))

m_lower = 1e7
m_upper = 1e10


plot_with_percentiles(
    ax,
    kbh_Repos1['BHMass'],
    kbh_Repos1['ReleasedKE'],
    xbins=30,
    color='red',
    xlower=m_lower,
    xupper=m_upper,
    label='Kn1-Repos1'
)
plot_with_percentiles(
    ax,
    kbh_DF['BHMass'],
    kbh_DF['ReleasedKE'],
    xbins=30,
    color='blue',
    xlower=m_lower,
    xupper=m_upper,
    label='Kn1-DF'
)


plot_with_percentiles(
    ax,
    kbh_SRepos1['BHMass'],
    kbh_SRepos1['ReleasedKE'],
    xbins=30,
    color='orange',
    xlower=m_lower,
    xupper=m_upper,
    label='Kn1S-Repos1'
)

plot_with_percentiles(
    ax,
    kbh_SDF['BHMass'],
    kbh_SDF['ReleasedKE'],
    xbins=30,
    color='green',
    xlower=m_lower,
    xupper=m_upper,
    label='Kn1S-DF'
)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel(r'BH Mass [M$_\odot/h$]')
ax.set_ylabel(r'Released KE per BH [$c^2$M$_\odot/h$]')
ax.legend()
plt.tight_layout()
plt.show()
```
