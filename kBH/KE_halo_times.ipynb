```python
# 1x2 figure for redshifts (zs has 2 elements)

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['text.usetex'] = True
# font
plt.rcParams["font.family"] = 'Nimbus Roman'
plt.rcParams['mathtext.fontset'] = 'cm'

# zs = np.array([0.0, 0.5, 1.0, 1.5])
zs = np.array([0.0, 1.0])

m_lower = 1e10
m_upper = 5e14
bins = 24
colors = ['blue', 'red', 'green', 'orange']
lstyles = ['-', '-', '--', '--']
sim_names = [r'$\textsc{SV-DF}$', r'$\textsc{SV-Repos}$', r'$\textsc{SV-DF-S}$', r'$\textsc{SV-Repos-S}$']

def sci_notation(x, digits=2):
    return f'{x:.{digits}e}'[:-4] + r"$\times 10^{" + str(int(f'{x:.{digits}e}'.split('e')[1])) + r"}$"

fig, axs = plt.subplots(2,1, figsize=(4., 5.9), sharex=True, sharey=True)
fig.subplots_adjust(wspace=0.03, hspace=0.03)
axs = axs.flatten()

for i, z in enumerate(zs):
    ax = axs[i]
    if z in kbh_DF:
        u_tot_DF = plot_energy_density_distribution(
            ax,
            kbh_DF[z],
            nbins=bins,
            color=colors[0],
            xlower=m_lower,
            xupper=m_upper,
            label=sim_names[0],
            ls=lstyles[0],
            masstype='M500c'
            )
    if z in kbh_Repos1:
        u_tot_R1 = plot_energy_density_distribution(
            ax,
            kbh_Repos1[z],
            nbins=bins,
            color=colors[1],
            xlower=m_lower,
            xupper=m_upper,
            label=sim_names[1],
            ls=lstyles[1],
            masstype='M500c'
            )
    if z in kbh_SDF:
        u_tot_SDF = plot_energy_density_distribution(
            ax,
            kbh_SDF[z],
            nbins=bins,
            color=colors[2],
            xlower=m_lower,
            xupper=m_upper,
            label=sim_names[2],
            ls=lstyles[2],
            masstype='M500c'
            )
    if z in kbh_SRepos1:
        u_tot_SR1 = plot_energy_density_distribution(
            ax,
            kbh_SRepos1[z],
            nbins=bins,
            color=colors[3],
            xlower=m_lower,
            xupper=m_upper,
            label=sim_names[3],
            ls=lstyles[3],
            masstype='M500c'
            )
        
    # annotate total energy density for each model with different colors
    ax.text(0.04, .92, 'Total energy density:', transform=ax.transAxes)  # placeholder for alignment
    if z in kbh_DF:
        ax.text(0.04, 0.86, sci_notation(u_tot_DF), transform=ax.transAxes, color=colors[0])
    if z in kbh_Repos1:
        ax.text(0.04, 0.80, sci_notation(u_tot_R1), transform=ax.transAxes, color=colors[1])
    if z in kbh_SDF:
        ax.text(0.04, 0.74, sci_notation(u_tot_SDF), transform=ax.transAxes, color=colors[2])
    if z in kbh_SRepos1:
        ax.text(0.04, 0.68, sci_notation(u_tot_SR1), transform=ax.transAxes, color=colors[3])
    

    # ax.set_title(f'z = {z:.1f}')
    # annotate z using text instead of title
    ax.text(0.79, 0.05, r'$z = %.1f$' % z, transform=ax.transAxes)
    ax.set_xscale('log')
    ax.set_yscale('log')
    
    # x limit
    ax.set_xlim(2*m_lower, .25*m_upper)
    ax.set_ylim(3e-2, .8e4)
    ax.grid(ls=":")
    
    if i == 1: # bottom panel
        ax.set_xlabel(r'$M_\mathrm{500c}$ [M$_\odot/h$]')
    
    # y label for all (since it's a column, but left side only matters if we had columns > 1. Here we have 1 column, so both rows get ylabel)
    # Actually, SKILL says "Keep y-axis labels and tick labels on the left column only"
    # Since we have nRows=2, nCols=1, all are in the "left column" (col 0).
    # So both should have ylabel? No, usually in vertical stack, we share x but keep y labels if scales are consistent?
    # SKILL says "For repeated quantities across panels... Keep y-axis labels... on the left column only."
    # Since we have col=0 for both, we keep it.
    
    # Wait, usually for stacked sharex plots:
    # Top panel: no xlabel. Bottom panel: xlabel.
    # Both panels: ylabel.
    
    ax.set_ylabel(r'$\mathcal{U}_\mathrm{kf}^\mathrm{halo}(M_\mathrm{500c},z)$ $\left[\frac{c^2\,\mathrm{M}_\odot\,h^{-1}}{ (\mathrm{cMpc}\,h^{-1})^{3}\,\mathrm{dex} }\right]$')

    ax.label_outer()
    
# legend only for the first subplot
axs[0].legend(loc='lower left', bbox_to_anchor=(.53, 0.13))
# plt.tight_layout()
# plt.show()
plt.savefig('fig_KE_density_M500c.pdf',  bbox_inches='tight')
```

```python
# plot feedback energy vs. halo mass

# we need to sum over BHs belonging to the same halo
def aggregate_ke_per_halo(kbh_data, halo_mass_type='M200c'):
    halo_ids = kbh_data['GroupID']
    halo_masses = kbh_data[halo_mass_type]
    released_ke = kbh_data['ReleasedKE']

    unique_halo_ids = np.unique(halo_ids)
    m_halo = []
    ke_halo = []

    for hid in unique_halo_ids:
        mask = halo_ids == hid
        total_ke = np.sum(released_ke[mask])
        halo_mass = halo_masses[mask][0]  # assuming all BHs in the same halo have the same halo mass
        m_halo.append(halo_mass)
        ke_halo.append(total_ke)

    return np.array(m_halo), np.array(ke_halo)

m_lower = 5e9
m_upper = 5e14

fig, axs = plt.subplots(2, 2, figsize=(9, 7))
axs = axs.flatten()

for i, z in enumerate(zs):
    if i >= 4:
        break
    ax = axs[i]
    if z in kbh_Repos1:
        m_halo, ke_halo = aggregate_ke_per_halo(kbh_Repos1[z])
        plot_with_percentiles(
            ax,
            m_halo,
            ke_halo,
            xbins=30,
            color=colors[0],
            xlower=m_lower,
            xupper=m_upper,
            label='Kn1-Repos1'
        )
    if z in kbh_DF:
        m_halo, ke_halo = aggregate_ke_per_halo(kbh_DF[z])
        plot_with_percentiles(
            ax,
            m_halo,
            ke_halo,
            xbins=30,
            color=colors[1],
            xlower=m_lower,
            xupper=m_upper,
            label='Kn1-DF'
        )
    if z in kbh_SRepos1:
        m_halo, ke_halo = aggregate_ke_per_halo(kbh_SRepos1[z])
        plot_with_percentiles(
            ax,
            m_halo,
            ke_halo,
            xbins=30,
            color=colors[2],
            xlower=m_lower,
            xupper=m_upper,
            label=f'Kn1S-Repos1',
            )
    if z in kbh_SDF:
        m_halo, ke_halo = aggregate_ke_per_halo(kbh_SDF[z])
        plot_with_percentiles(
            ax,
            m_halo,
            ke_halo,
            xbins=30,
            color=colors[3],
            xlower=m_lower,
            xupper=m_upper,
            label=f'Kn1S-DF'
            )

    # ax.set_title(f'z = {z:.1f}')
    # annotate z using text instead of title
    ax.text(0.45, 0.05, f'z = {z:.1f}', transform=ax.transAxes)
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel(r'$M_\mathrm{200c}$ [M$_\odot/h$]')
    if i % 2 == 0:
        ax.set_ylabel(r'KE [$c^2$M$_\odot/h$]')
    # ax.legend()
    # x limit
    ax.set_xlim(m_lower, m_upper)
    ax.set_ylim(1e2, 1e8)
# legend only for the first subplot
axs[0].legend(loc='upper left')
plt.tight_layout()
plt.show()
```
