```python
import glob
import numpy as np 
import os
import matplotlib.pyplot as plt
from utils import *
```

```python
# load data
zs = np.array([0.0, 0.2, 0.5, 1.0, 2.0])

sim_names = ['kn1-Repos1', 'kn1-DF', 'kn1S-Repos1', 'kn1S-DF']

kbh_Repos1 = get_kBH_end_zs(sim_names[0], zs, halo_info=True)
kbh_DF = get_kBH_end_zs(sim_names[1], zs, halo_info=True)
kbh_SRepos1 = get_kBH_end_zs(sim_names[2], zs, halo_info=True)
kbh_SDF = get_kBH_end_zs(sim_names[3], zs, halo_info=True)
```

```python
# 2x2 figure for redshifts

import matplotlib as mpl
mpl.rcParams['text.usetex'] = True
# font
plt.rcParams['font.family'] = 'Nimbus Roman'
plt.rcParams['mathtext.fontset'] = 'cm'

m_lower = 5e9
m_upper = 5e14
colors = ['red', 'blue', 'orange', 'green']

nrows, ncols = 2, 2
fig, axs = plt.subplots(nrows, ncols, figsize=(9, 7))
fig.subplots_adjust(wspace=0.2, hspace=0.2)
axs = axs.flatten()

for i, z in enumerate(zs):
    if i >= 4:
        break
    ax = axs[i]
    if z in kbh_Repos1:
        u_tot_R1 = plot_energy_density_distribution(
            ax,
            kbh_Repos1[z],
            nbins=25,
            color=colors[0],
            xlower=m_lower,
            xupper=m_upper,
            label=f'Kn1-Repos1',
            linewidth=2.0,
            masstype='M200c'
            )
    if z in kbh_DF:
        u_tot_DF = plot_energy_density_distribution(
            ax,
            kbh_DF[z],
            nbins=25,
            color=colors[1],
            xlower=m_lower,
            xupper=m_upper,
            label=f'Kn1-DF',
            linewidth=2.0,
            masstype='M200c'
            )
    if z in kbh_SRepos1:
        u_tot_SR1 = plot_energy_density_distribution(
            ax,
            kbh_SRepos1[z],
            nbins=25,
            color=colors[2],
            xlower=m_lower,
            xupper=m_upper,
            label=f'Kn1S-Repos1',
            linewidth=2.0,
            masstype='M200c'
            )
    if z in kbh_SDF:
        u_tot_SDF = plot_energy_density_distribution(
            ax,
            kbh_SDF[z],
            nbins=25,
            color=colors[3],
            xlower=m_lower,
            xupper=m_upper,
            label=f'Kn1S-DF',
            linewidth=2.0,
            masstype='M200c'
            )
    # annotate total energy density for each model with different colors
    ax.text(0.05, .95, 'Total energy density:', transform=ax.transAxes)  # placeholder for alignment
    if z in kbh_Repos1:
        ax.text(0.05, 0.9, f'{u_tot_R1:.2e}', transform=ax.transAxes, color=colors[0])
    if z in kbh_DF:
        ax.text(0.05, 0.85, f'{u_tot_DF:.2e}', transform=ax.transAxes, color=colors[1])
    if z in kbh_SRepos1:
        ax.text(0.05, 0.8, f'{u_tot_SR1:.2e}', transform=ax.transAxes, color=colors[2])
    if z in kbh_SDF:
        ax.text(0.05, 0.75, f'{u_tot_SDF:.2e}', transform=ax.transAxes, color=colors[3])

    # ax.set_title(f'z = {z:.1f}')
    # annotate z using text instead of title
    ax.text(0.45, 0.05, f'z = {z:.1f}', transform=ax.transAxes)
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim(m_lower, m_upper)
    ax.set_ylim(5e-3, 1e4)
    ax.grid(ls=':')

    row = i // ncols
    col = i % ncols
    if row == nrows - 1:
        ax.set_xlabel(r'$M_\mathrm{200c}$ [M$_\odot/h$]')
    else:
        ax.set_xticklabels([])
    if col == 0:
        ax.set_ylabel(r'KE Density [$c^2h^{-1}$M$_\odot$(Mpc/h)$^{-3}\,\mathrm{dex}^{-1}$]')
    else:
        ax.set_yticklabels([])

# legend only for the first subplot
axs[0].legend(loc='center left', bbox_to_anchor=(.57, 0.5))
plt.tight_layout()
plt.show()
```
